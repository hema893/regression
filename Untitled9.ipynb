````python
Question 1: What is Simple Linear Regression?
````

````python
**Simple Linear Regression** is a statistical method used to model the relationship between **one independent variable (X)** and **one dependent variable (Y)** by fitting a straight line through the data.
The goal is to **predict** Y based on X using the equation:

$$
Y = \beta_0 + \beta_1 X + \varepsilon
$$

Where $\beta_0$ is the intercept, $\beta_1$ is the slope, and $\varepsilon$ is the error term.
````

````python
Question 2: What are the key assumptions of Simple Linear Regression?
````

````python
The key assumptions of **Simple Linear Regression** are:

1. **Linearity**: The relationship between X and Y is linear.
2. **Independence**: Residuals (errors) are independent of each other.
3. **Homoscedasticity**: Residuals have constant variance across all values of X.
4. **Normality**: Residuals are normally distributed.
````

````python
Question 3: What is heteroscedasticity, and why is it important to address in regression models?
````

````python
**Heteroscedasticity** occurs when the **variance of residuals (errors)** is **not constant** across all levels of the independent variable.

### üîç Why it matters:

* It **violates regression assumptions**
* Leads to **biased standard errors**
* Makes **hypothesis tests and confidence intervals unreliable**
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate data with heteroscedasticity
np.random.seed(0)
X = np.linspace(1, 10, 100).reshape(-1, 1)
y = 2 * X.squeeze() + np.random.randn(100) * X.squeeze()  # error increases with X

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
residuals = y - y_pred

# Plot residuals to show heteroscedasticity
plt.figure(figsize=(8, 5))
plt.scatter(X, residuals, color='purple', alpha=0.7)
plt.axhline(0, color='black', linestyle='--')
plt.title("Residual Plot Showing Heteroscedasticity")
plt.xlabel("X")
plt.ylabel("Residuals")
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
Question 4: What is Multiple Linear Regression?
````

````python
**Multiple Linear Regression** is a statistical method used to model the relationship between **one dependent variable** and **two or more independent variables**.

### Equation:

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_n X_n + \varepsilon
$$

It helps in **predicting** outcomes and understanding the **impact of multiple factors** on a single response variable.
````

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Sample data
np.random.seed(1)
X1 = np.random.rand(50) * 10
X2 = np.random.rand(50) * 10
Y = 3 * X1 + 2 * X2 + 5 + np.random.randn(50) * 2  # Y depends on X1 and X2

# Combine X1 and X2 into a feature matrix
X = np.column_stack((X1, X2))

# Fit the Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Plotting in 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, Y, color='blue', label='Actual Data')
ax.scatter(X1, X2, Y_pred, color='red', label='Predicted Plane')

ax.set_title('Multiple Linear Regression (3D View)')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
Question 5: What is polynomial regression, and how does it differ from linear regression?
````

````python
**Polynomial Regression** is a type of regression where the relationship between the independent variable $X$ and the dependent variable $Y$ is modeled as an **nth-degree polynomial**.

### Example:

$$
Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \dots + \beta_n X^n + \varepsilon
$$

### üîÑ Difference from Linear Regression:

* **Linear Regression** fits a **straight line** to the data.
* **Polynomial Regression** fits a **curved line**, capturing **non-linear relationships** between variables.
````

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 3 * X.squeeze()**2 - 2 * X.squeeze() + 1 + np.random.randn(50) * 10

# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_lin = lin_reg.predict(X)

# Polynomial Regression (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
poly_reg = LinearRegression()
poly_reg.fit(X_poly, y)
y_poly = poly_reg.predict(X_poly)

# Plotting
plt.figure(figsize=(8, 5))
plt.scatter(X, y, color='gray', label='Data Points')
plt.plot(X, y_lin, color='blue', label='Linear Regression')
plt.plot(X, y_poly, color='red', label='Polynomial Regression')
plt.title('Linear vs Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("regression_chart.png")  # Save the chart
plt.show()
````

````python
Question 7: Fit a Multiple Linear Regression model on this sample data:
‚óè Area = [1200, 1500, 1800, 2000]
‚óè Rooms = [2, 3, 3, 4]
‚óè Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.
(Include your Python code and output in the code box below?
````

````python
Here‚Äôs a complete Python program to:

Fit a Multiple Linear Regression model

Check for Multicollinearity using Variance Inflation Factor (VIF)

````

````python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[['Area', 'Rooms']]
y = df['Price']

# Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, y)

# Print coefficients
print(f"Intercept: {model.intercept_}")
print(f"Coefficients: {model.coef_}")

# Add constant for VIF calculation
X_with_const = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
````

````python
Question 8: Implement polynomial regression on the following data:
‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]
Fit a 2nd-degree polynomial and plot the resulting curve.
````

````python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Fit a 2nd-degree polynomial
coeffs = np.polyfit(X, Y, 2)
poly = np.poly1d(coeffs)

# Generate X values for the smooth curve
X_curve = np.linspace(min(X), max(X), 100)
Y_curve = poly(X_curve)

# Plotting
plt.scatter(X, Y, color='red', label='Original Data')
plt.plot(X_curve, Y_curve, color='blue', label=f'2nd-degree Fit:\n$y = {coeffs[0]:.2f}x^2 + {coeffs[1]:.2f}x + {coeffs[2]:.2f}$')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2nd-Degree Polynomial Regression')
plt.legend()
plt.grid(True)
plt.show()
````

````python
Question 9: Create a residuals plot for a regression model trained on this data:
‚óè X = [10, 20, 30, 40, 50]
‚óè Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.
````

````python
Here‚Äôs a complete **Python program** to:

‚úÖ Fit a **Simple Linear Regression** model
‚úÖ Create a **residuals plot**
‚úÖ Visually assess **heteroscedasticity**

---

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit the regression model
model = LinearRegression()
model.fit(X, Y)
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.figure(figsize=(8, 5))
plt.scatter(Y_pred, residuals, color='purple', marker='o')
plt.axhline(0, color='black', linestyle='--')
plt.title('Residuals Plot')
plt.xlabel('Predicted Y')
plt.ylabel('Residuals')
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

### ‚úÖ How to Interpret:

* If residuals are **evenly spread around 0**, there is **no heteroscedasticity**.
* If the spread **increases or decreases with X or predicted Y**, it indicates **heteroscedasticity**.

Let me know if you want statistical tests (like Breusch‚ÄìPagan) to confirm heteroscedasticity!
````

````python
As a data scientist facing **heteroscedasticity** and **multicollinearity** in a regression model for predicting house prices, here are the steps I would take to ensure a **robust and reliable model**:

---

### üîç **1. Addressing Heteroscedasticity**

Heteroscedasticity means that the **variance of residuals changes across levels of predictors**, violating a key regression assumption.

#### ‚úÖ **Steps to handle it:**

* **Log or Power Transformation**
  Apply a log transformation to the dependent variable (e.g., `log(Price)`) or predictors:

  ```python
  df['log_price'] = np.log(df['Price'])
  ```

* **Use Weighted Least Squares (WLS)**
  Assign lower weights to observations with higher variance.

* **Use Robust Standard Errors**
  Switch to methods like **Huber-White robust standard errors** using statsmodels:

  ```python
  model = sm.OLS(y, X).fit(cov_type='HC3')
  ```

---

### üîÑ **2. Addressing Multicollinearity**

Multicollinearity means that **predictor variables are highly correlated** with each other, which can make coefficient estimates unstable.

#### ‚úÖ **Steps to handle it:**

* **Calculate VIF (Variance Inflation Factor)**
  Drop or combine variables with **high VIF (>5 or 10)**.

  ```python
  from statsmodels.stats.outliers_influence import variance_inflation_factor
  ```

* **Remove or Combine Correlated Features**
  If `Area` and `Rooms` are highly correlated, consider removing one or combining them into a new feature (e.g., `Area per Room`).

* **Apply Regularization Techniques**
  Use **Ridge** or **Lasso regression** to reduce the impact of multicollinearity.

  ```python
  from sklearn.linear_model import Ridge, Lasso
  ```

---

### üèóÔ∏è **3. Model Improvement and Validation**

* **Feature Engineering**
  Include categorical variables like `Location` using **one-hot encoding**.

* **Cross-Validation**
  Use **k-fold cross-validation** to evaluate model stability.

* **Residual Analysis**
  Always visualize residual plots to check for remaining issues.

---

### ‚úÖ Final Outcome:

By applying transformations, robust methods, and regularization, you build a model that:

* Produces **reliable coefficient estimates**
* Meets **regression assumptions**
* Improves **prediction accuracy** for house price
````
